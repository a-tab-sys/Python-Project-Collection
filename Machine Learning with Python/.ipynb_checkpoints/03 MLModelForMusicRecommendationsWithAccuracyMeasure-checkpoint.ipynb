```python
# Problem: We have an online music store
# When users sign up they are asked to provide an age and gender
# Based on their profile i want to recomment various music albums they are likely to buy
# We want to use ML to increase sales
# Want to build a model. Feed model with sample data based on existing users
# Our models will learn the patterns in our data so we can ask it to make predictions
```

```python
# 1. Import Data
# 2. Clean Data
# 3. Slpil Data into training/test sets
# 4. Create a model
# 5. Train the model
# 6. Make Predictions
# 7. Evaluate and improve
```

```python
import pandas as pd
# music_data is a dataframe object
music_data=pd.read_csv('03 music.csv')
music_data
```

```python
# Clean the data - already clean, no null values.
# We should split the data into input (age, gender) and output (genre)
# When we train a model, we need to provide it with an input dataset and an output dataset
```

```python
# Creating input set
# Shift+tab with cursor on drop to see details. Drop does not modify original dataset, it will create a new data set, removing specified rows or coulmns  
# By convention, capital x is used to represent the input dataset
X = music_data.drop(columns=['genre'])
X
```

```python
# Creating output set
y=music_data['genre']
y
```

```python
# Split dataset into 2 sets, one for training and one for testing
from sklearn.model_selection import train_test_split
# train_test_split function takes 3 arguments <outputs, inputs, keyword arg that specifies the size of our test set>
# Function returns a tupple so we can unpack it into 4 variables. First 2 variables are input sets for training & testing. Last 2 variables are  output sets for training and testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #allocating about 20% of our data for testing
```

```python
# Create a model using a ML algorithm
# We are going to use the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier

# Create a new instance of this class
model=DecisionTreeClassifier()
```

```python
# Train the model so it learnpatterns in the data
# Method takes 2 data sets - input and output sets
model.fit(X_train, y_train)
```

```python
# Ask model to make predictions. Use iputs are are not apart of original training dataset
# Method takes a 2-Dimensional array.
# predictions=model.predict([[21,1], [22, 0]])  #Passing 21 year old male, and 22 year old female
predictions = model.predict(X_test)
predictions
```

```python
#Predicted that 21 male like hiphop
#Predicted that 22 female like dance music
```

```python
# Split dataset into 2 sets, one for training and one for testing
# We had used 2 samples to test prediction, that is not enough

# General rule of thumb- allocate 70-80% of our data for training and the other 20-30% for testing
# Then instead of passing only 2 samples for making predictions, we can pass the dataset we have for testing and get the predictions. 
# Compare predictions with values from test set, based of that calculate the accuracy
```

```python
# Measure accuracy of your model
# compare predictions to y_test (test outputs)
from sklearn.metrics import accuracy_score
# accuracy score funtion outputs a number between 0-1 to represent accuracy
score = accuracy_score(y_test, predictions)
score
```
